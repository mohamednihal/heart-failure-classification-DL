```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from collections import Counter
from sklearn.compose import ColumnTransformer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, InputLayer
from sklearn.metrics import classification_report
from tensorflow.keras.utils import to_categorical
import numpy as np
#read the data

data = pd.read_csv('heart_failure.csv')
print(data.info())

print('Classes and number of value in the dataset', Counter(data['death_event']))


# print(data['DEATH_EVENT'])
# extracting label
y = data['death_event']
#extracting fatures
x = data[['age','anaemia','creatinine_phosphokinase','diabetes','ejection_fraction','high_blood_pressure','platelets','serum_creatinine','serum_sodium','sex','smoking','time']]
#one hot encoding the features

x = pd.get_dummies(x)

#split the data
X_train,X_test,Y_train,Y_test = train_test_split(x, y , test_size = 0.2, random_state = 42)

#transferring columns
ct = ColumnTransformer([('standardscaler', StandardScaler(),['age','creatinine_phosphokinase','ejection_fraction','platelets','serum_creatinine','serum_sodium','time'])])

#train
X_train = ct.fit_transform(X_train)
#test
X_test = ct.transform(X_test)


#label encoder for classification
le = LabelEncoder()
#training labels
Y_train = le.fit_transform(Y_train.astype(str))
#testing labels 
Y_test  = le.transform(Y_test.astype(str))

#converting encoded labels into binary vectors
Y_train =to_categorical(Y_train)
#
Y_test = to_categorical(Y_test)

#initialize the tensorflow model
model = Sequential()
#input layer
model.add(InputLayer(input_shape=(X_train.shape[1],)))
#hidden layer
model.add(Dense(12, activation = 'relu'))
#output layer
model.add(Dense(2,activation = 'softmax'))
#compile 
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
#fit the model
model.fit(X_train,Y_train,epochs = 100,batch_size = 2 , verbose = 1)

#loss and actual metrics
loss, acc = model.evaluate(X_test,Y_test, verbose= 0)

#predicting
y_estimate = model.predict(X_test, verbose = 0)
y_estimate = np.argmax(y_estimate, axis=1)
# true
y_true = np.argmax(Y_test, axis =1)
print(classification_report(y_true, y_estimate))


```
